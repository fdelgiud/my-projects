```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
urls = [
    'https://it.wikipedia.org/wiki/Grammy_Awards_2014',
    'https://it.wikipedia.org/wiki/Grammy_Awards_2015',
    'https://it.wikipedia.org/wiki/Grammy_Awards_2016',
    'https://it.wikipedia.org/wiki/Grammy_Awards_2017',
    'https://it.wikipedia.org/wiki/Grammy_Awards_2018',
    'https://it.wikipedia.org/wiki/Grammy_Awards_2019',
    'https://it.wikipedia.org/wiki/Grammy_Awards_2020',
    'https://it.wikipedia.org/wiki/Grammy_Awards_2021',
    'https://it.wikipedia.org/wiki/Grammy_Awards_2022',
    'https://it.wikipedia.org/wiki/Grammy_Awards_2023',
    'https://it.wikipedia.org/wiki/Grammy_Awards_2024',
]

all_dfs = []

for url in urls:
    year = url.split("_")[-1]
    response = requests.get(url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, "html.parser")
        ul_elements = soup.find_all("ul")
        categories = []
        winners = []
       
        for ul_element in ul_elements:
            li_elements = ul_element.find_all("li")
           
            for li_element in li_elements:
                winner = li_element.get_text(strip=True)
               
                bold_text = li_element.find("b")
                if bold_text:
                    category = bold_text.find_previous("span", {"class": "mw-headline"}).get_text(strip=True) if bold_text.find_previous("span", {"class": "mw-headline"}) else ""
                   
                    categories.append(category)
                    winners.append(winner)
       
        df = pd.DataFrame({
            "Categoria": categories,
            "Winner": winners,
            "YearOfCeremony": [year] * len(categories)})
        all_dfs.append(df)
    else:
        print(f"Errore nella richiesta HTTP per l'URL: {url}")

GrammyAwards = pd.concat(all_dfs, ignore_index=True)
GrammyAwards.head(10)
```

```python
GrammyAwards['YearOfCeremony'] = GrammyAwards['YearOfCeremony'].astype(int)
GrammyAwards["Year"] = GrammyAwards["YearOfCeremony"]-1
GrammyAwards['Competition'] = 'GrammyAwards'
```

```python
def split_artists(df, column_name):
    # Contatore per tenere traccia del numero di artisti
    artist_count = 1
    
    # Itera su ogni riga del DataFrame
    for index, row in df.iterrows():
        # Controlla se il valore nella colonna degli artisti non è None
        if pd.notna(row[column_name]):
            # Dividi i nomi degli artisti utilizzando virgola, "and", e "with" come separatori
            artists = re.split(r'[,]|(?:\sand\s)|\swith\s', row[column_name])
            # Aggiungi i nomi degli artisti come nuove colonne nel DataFrame
            for i, artist in enumerate(artists, start=1):
                df.at[index, f'Artist {i}'] = artist.strip()

    # Rimuovi la colonna originale degli artisti
    df.drop(columns=[column_name], inplace=True)
    
    return df


GrammyAwards = split_artists(GrammyAwards, 'Artist')
```

# Analysis

### Who is the most successful Artist?

```python
from collections import Counter

# Rimuovi i valori nulli prima di unire le colonne degli artisti in un'unica lista
artists_list = [artist for artist in GrammyAwards_final[['Artist 1', 'Artist 2']].values.ravel() if pd.notnull(artist)]

# Conta le occorrenze di ogni artista utilizzando Counter
artist_counter = Counter(artists_list)

# Trova l'artista con il massimo numero di occorrenze
most_successful_artist, most_successful_artist_count = artist_counter.most_common(1)[0]

print("The most successful Artist is:", most_successful_artist)
print("Number of Awards:", most_successful_artist_count)
```

### Which is the most awarded Album?
