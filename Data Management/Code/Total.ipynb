```python
import pandas as pd
```

```python
AmericanAwards = pd.read_csv('AMA_final.csv')
BillboardAwards = pd.read_csv('BMA_final.csv')
GrammyAwards = pd.read_csv('GA_final.csv')
```

```python
merged_df = pd.concat([AmericanAwards, BillboardAwards, GrammyAwards], axis=0)
merged_df = merged_df.drop('Unnamed: 0', axis=1)
print(merged_df.info())
```

```python
merged_df.to_csv('Total.csv')
```

## ALBUM

```python
mapping = {
    'Best Latin Pop Album': 'Top Latin Album',
    'Best R&B Album': 'Top R&B Album',
    'Best Rap Album': 'Top Rap Album',
    'Best Traditional Pop Album': 'Top Pop Album',
    'Best Rock Album': 'Top Rock Album',
    'Favorite Album - Latin': 'Top Latin Album',
    'Favorite Album - R&B': 'Top R&B Album',
    'Favorite Album - Rap/Hip-Hop': 'Top Rap Album',
    'Favorite Album - Pop': 'Top Pop Album',
    'Favorite Album - Rock': 'Top Rock Album',
    'Top Billboard 200 Album':'Top Album',
    'Album Of The Year':'Top Album',
    'Favorite Album - Soul/R&B':'Top R&B Album',
    'Best Latin Rock Album': 'Top Latin Album',
    'Best Latin Or Latin Alternative Latin Album': 'Top Latin Album',
    'Best Dance/Electronic Album':'Top Dance/Electronic Album',
    'Top EDM Album':'Top Dance/Electronic Album'
}


merged_df['Category'] = merged_df['Category'].apply(lambda x: mapping[x] if x in mapping else x)
```

```python
import pandas as pd

categorie_desiderate = ['Top Latin Album', 'Top R&B Album', 'Top Rap Album', 'Top Pop Album', 'Top Rock Album','Top Album','Top Dance/Electronic Album']
filtered_df = merged_df[merged_df['Category'].isin(categorie_desiderate)]

filtered_df.loc[:, 'Album'] = filtered_df['Album'].astype(str)

years = filtered_df['Year'].unique()
categories = categorie_desiderate
competitions = ['AmericanMusicAwards', 'BillboardMusicAwards', 'GrammyAwards']

all_combinations = []
for year in years:
    for category in categories:
        for competition in competitions:
            all_combinations.append((year, category, competition))

all_combinations_df = pd.DataFrame(all_combinations, columns=['Year', 'Category', 'Competition'])

merged_contest = pd.merge(all_combinations_df, filtered_df, on=['Year', 'Category', 'Competition'], how='left')

grouped_df = merged_contest.groupby(['Year', 'Category', 'Competition'])['Album'].apply(lambda x: ', '.join(x.unique().astype(str))).reset_index()

winner_ama = grouped_df[grouped_df['Competition'] == 'AmericanMusicAwards'].set_index(['Year', 'Category'])['Album'].rename('Winner_AMA')
winner_bma = grouped_df[grouped_df['Competition'] == 'BillboardMusicAwards'].set_index(['Year', 'Category'])['Album'].rename('Winner_BMA')
winner_ga = grouped_df[grouped_df['Competition'] == 'GrammyAwards'].set_index(['Year', 'Category'])['Album'].rename('Winner_GA')

Album_contest = winner_ama.to_frame().join(winner_bma).join(winner_ga)

Album_contest.reset_index(inplace=True)
```
